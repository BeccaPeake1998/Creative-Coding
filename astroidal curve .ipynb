```python
import matplotlib.pyplot as plt
import numpy as np


#create some points

t=np.linspace(-10*np.pi,10*np.pi,100)

fig,axes = plt.subplots(nrows=3, ncols=3)
#set up the graph


for n in range(9): 
        a=t**n
        x=a*np.cos(t)**3  
        y=a*np.sin(t)**3

  #the graph
        row,col=np.indices((3,3))
        r = row.ravel()[n]
        c = col.ravel()[n]
        
        axes[r,c].plot(x, y, color = 'white')
        axes[r,c].axis('off')
        #axes[r,c].axis('square')
        
plt.style.use('dark_background')    
plt.show()

```
